```python
def dictionary_maker(list_):
    subject_letter_amount = {} # Creates an empty dictionary
    for word in list_:
        subject_letter_amount[word] = { 
            "Length of word" : len(word), "Even or Odd?" : "Even" if len(word)%2 == 0 else "Odd"
        } # The for loop adds elements of the list to the dictionary and 
          #  assigns % values
    return subject_letter_amount    
list_of_strings = ["Math", "Science", "English", "Social"]
print(dictionary_maker(list_of_strings))
```
