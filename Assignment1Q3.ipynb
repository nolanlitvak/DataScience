```python
def power(x, y):
    answer = x**y
    return answer # 
list_of_pairs = [(9, 1), (3, -2), (4, 2), (3, 4), (6, 3), (2, 5), (7, -4)]
final_answers = []
for x, y in list_of_pairs: # the for loop ignores negative y values
    if y < 0:
        continue
    final_answers.append(power(x, y))
    
print(f"The results are {final_answers}.") 
```
