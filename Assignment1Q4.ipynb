```python
from random import random
values = [random() for i in range(20)] # This creates 20 different 
x = random()                           # values between 0 and 1.
values.sort() # Sorts the values from smallest to largest
print(f"The ordered list of values are {values}.")
print(f"The x value is {x}.")
greater_values = []
for decimal in values:
    if x <= decimal:
        greater_values.append(decimal) # Adds values to new list that 
                                       # are greater or equal to x.
print(f"The first matching index is {greater_values[0]}.")
        
```
