```python
def circleAreaCoverage(radiusOfCircle1, radiusOfCircle2):
    if radiusOfCircle1 <= 0 or radiusOfCircle2 <= 0:
        return "The radius has to be a positive integer!"
    radiusOfCircle1 = int(radiusOfCircle1)
    radiusOfCircle2 = int(radiusOfCircle2)
    pi = 3.14159265
    area_of_circle_1 = pi * radiusOfCircle1**2
    area_of_circle_2 = pi * radiusOfCircle2**2
    if area_of_circle_1 >= area_of_circle_2:
        denominator = area_of_circle_1
        numerator = area_of_circle_2
    else:
        denominator = area_of_circle_2
        numerator = area_of_circle_1
    print(f"The area of Circle 1 is {area_of_circle_1}.")
    print(f"The area of Circle 2 is {area_of_circle_2}.")
    return (numerator / denominator) * 100

print(f"The percentage of covered area in the biggest circle is {circleAreaCoverage(2, 5)}%.")
```
