```python
def distribution_analysis(NumberEvaluater):
    ValueDictionary = {}
    number_amount = len(NumberEvaluater)
    for value in NumberEvaluater: 
        if value not in ValueDictionary:
            count = 0
            for Number in NumberEvaluater: # Increases the count when 
                if Number <= value:  # the number is less/= to value
                    count += 1
                ValueDictionary[value] = (count / number_amount) * 100
    return dict(sorted(ValueDictionary.items()))
    
    


numbers = [9, 4, 2, 7, 8, 3]
print(distribution_analysis(numbers))
```
