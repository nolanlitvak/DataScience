```python
def seconds_converter(seconds):
    if seconds < 0:
        return "Invalid! Seconds have to be a positive integer."
    hour = seconds // 3600   # Ensures proper units when counting 
    minute = (seconds%3600) // 60 # minutes hours and seconds.
    second = seconds % 60  # This makes sure the seconds arent more than
    if hour < 12:          # 60 when counting. 
        time_period = "AM"
    else:
        time_period = "PM"
    return f"Since midnight, it's been {hour} hours, {minute} minutes, and {second} seconds."
print(seconds_converter(15237))
```
